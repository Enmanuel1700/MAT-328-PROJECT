```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nba_df = pd.read_csv("https://raw.githubusercontent.com/Enmanuel1700/MAT-328-PROJECT/main/NBA_most_pay.csv")
nba_df
```

```python
nba_df.describe()  #used to view some basic statistical details like percentile, mean, std, etc.
```

```python
nba_df.dtypes #To get the data types of the data frame
```

```python
nba_df.isna().sum() #There is not missing or wrong data in the df
```

```python
nba_df.rename(columns = {"Unnamed: 0":"Raking"}, inplace = True) #Replacing the first column name Unnamed for Raking 
nba_df.rename(columns = {" Salary ":"Salary"}, inplace = True)
nba_df
```

```python
nba_df['Salary'] = nba_df['Salary'].replace('[$,]', '', regex=True).astype(float)
nba_df
```

```python
# Assuming 'nba' is your data frame
selected_columns = ['Age', 'GP', 'PTS', 'Salary']

# Plotting the distributions
for column in selected_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(nba_df[column], bins=20)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    
```

```python
nba_df_top=nba_df.loc[0:9,["Player Name","Raking"]] #Showing top ten most pay 
nba_df_top
```

```python
# Sort the DataFrame by salary in descending order
df_sorted = nba_df.sort_values(by='Salary', ascending=False)

# Select the top ten players
top_ten_players = df_sorted.head(10)
df_sorted.head(10)
```

```python
# Assuming 'nba' is your data frame
selected_columns_relationships = [('Age', 'PTS'), ('Salary', 'PTS')]

# Plotting the relationships
for x_column, y_column in selected_columns_relationships:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=nba_df, x=x_column, y=y_column, color='coral')
    plt.title(f'Relationship between {x_column} and {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.grid(True)
    plt.show()
```
